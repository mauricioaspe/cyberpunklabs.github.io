```python
BD = pd.read_excel ('/home/diogenes/projects/cyberpunklabs.github.io/scripts/basesDatos/hortaliceres_nn.xlsx',
                    decimal=',') #for an earlier version of Excel, you may need to use the file extension of 'xls'


BD.columns = ['productor', 'direccion', 'telefono', 'geolocalizacion']
BD.head()
```

```python
ruta_archivo = '/home/diogenes/projects/cyberpunklabs.github.io/scripts/basesDatos/hortaliceres_nn.datos'
icono = 'lechugaIcon'
sector = 'hortaliceres'

with open(ruta_archivo, 'w') as archivo_temporal:
    for productor in range(len(BD.productor)):
        entrada = u'// {}\nL.marker([{}], {{icon: {}}}).bindPopup(\'{}<br>{}\').addTo({});\n\n'\
        .format(BD.productor[productor].title(),
                BD.geolocalizacion[productor],
                icono,
                BD.direccion[productor].title(),
                BD.telefono[productor],
                sector) #.encode('utf-8').strip()
        
        archivo_temporal.write(entrada + '\n\n')

        
        
#L.marker([-32.4532208, -71.2499938], {icon: lechugaIcon}).bindPopup('This is Littleton, CO.').addTo(cities1),
```
